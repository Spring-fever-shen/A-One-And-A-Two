### 查看数据编码

```python
import chardet
#导入库，这是一个专门用于检测文件编码的工具，能自动分析文件内容并判断其编码格式。
def check_encoding(filename):
    rawdata = open(filename, 'rb').read()
    result = chardet.detect(rawdata)
    encoding = result['encoding']
    #从结果字典中提取编码格式
    confidence = result['confidence']
    #从结果字典中提取可信度
    return encoding, confidence

file_path = '000001.csv'# 添加文件路径
encoding, confidence = check_encoding(file_path)
print(f"Encoding: {encoding}, Confidence: {confidence}")
```

#### 按照时间

```python
data['Day'] = pd.to_datetime(data['Day'],format = '%Y/%m/%d')
data
# 将'Day'列从字符串格式转换为日期时间格式
# 将 DataFrame 中 Day 列的文本格式日期转换为 pandas 的 datetime 类型
# format='%Y/%m/%d' 指定了输入日期的格式：年/月/日
# 其中%Y、%m 等格式符的大小写也有区别，大写 Y 表示 4 位年份（如 2023）；小写 %y 表示 2 位年份
```

```python
data = data.sort_values(by=['Day'], axis=0, ascending=False)
data
# 按照 Day 列对 DataFrame 进行排序
# ascending=False：设置排序方向为降序（从大到小）如果设为True，则是升序
```
